```
import urllib.request as urllib2
import json
import csv
from urllib.parse import unquote,quote


MainUrl="https://misbar.com/v1/factcheck/listArticles?language=Arabic"
pages = {"misleading","commotion","false","selective","unclear","true","myth","satire"}

for x in pages:
  count=0
  csvfile = open('drive/My Drive/Arabic Fake news/Misbar/csvfiles/'+x+'.csv', 'w+', newline='')
  writer = csv.writer(csvfile)
  for i in range(100):
    URL = MainUrl+"&page="+str(i)+"&limit=15&scat=&result="+x
    response = urllib2.urlopen(URL)
    data = json.loads(response.read())
    if len(data) !=0 :
      for item in data['data']:
        if item['url_path']:
          href="https://misbar.com/"+quote(item['url_path'])
          writer.writerow([href])
          count=count+1
  print(str(x)+".csv count = "+str(count))
  csvfile.close()

print("Finish")
```

```
from io import StringIO
from html.parser import HTMLParser

class MLStripper(HTMLParser):
    def __init__(self):
        super().__init__()
        self.reset()
        self.strict = False
        self.convert_charrefs= True
        self.text = StringIO()
    def handle_data(self, d):
        self.text.write(d)
    def get_data(self):
        return self.text.getvalue()

def strip_tags(html):
    s = MLStripper()
    s.feed(html)
    return s.get_data()
```

```
import urllib.request as urllib2
from urllib.parse import unquote,quote
import xlsxwriter
import json
import os
import csv

pages = {
  "misleading": 'drive/My Drive/Arabic Fake news/Misbar/csvfiles/misleading.csv',
  "commotion": 'drive/My Drive/Arabic Fake news/Misbar/csvfiles/commotion.csv',
  "false":'drive/My Drive/Arabic Fake news/Misbar/csvfiles/false.csv',
  "selective": 'drive/My Drive/Arabic Fake news/Misbar/csvfiles/selective.csv',
  "unclear": 'drive/My Drive/Arabic Fake news/Misbar/csvfiles/unclear.csv',
  "true":'drive/My Drive/Arabic Fake news/Misbar/csvfiles/true.csv',
  "myth": 'drive/My Drive/Arabic Fake news/Misbar/csvfiles/myth.csv',
  "satire":'drive/My Drive/Arabic Fake news/Misbar/csvfiles/satire.csv'}

for x in pages:
  basePath='drive/My Drive/Arabic Fake news/Misbar/XlsxData/OriginalData/'+str(x)+'.xlsx'
  workbook = xlsxwriter.Workbook(basePath)
  worksheet = workbook.add_worksheet()

  file = open(pages[x])
  csv_file = csv.reader(file)
  i=0 

  for link in csv_file:
    link='/'.join(link[0].split("/")[-4:])
    link='https://misbar.com/v1/factcheck/getBySlug/'+str(link)
    response = urllib2.urlopen(link)
    data = json.loads(response.read())
    c_data=data['data']

    claim=strip_tags(c_data["claim"])
    title=c_data["title"]
    category=c_data["category"]
    #caption=c_data["caption"]
    url_path="https://misbar.com/"+quote(c_data['url_path'])

    recored=[title,claim,category,url_path]

    worksheet.write_row(i, 0, recored)
    i=i+1
  workbook.close()
  print("Finish Collecting " +str(x)+" "+str(i) +" news ") 

```

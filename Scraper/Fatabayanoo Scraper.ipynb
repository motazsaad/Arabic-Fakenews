# Collect urls

```
import requests
from bs4 import BeautifulSoup
import csv

pages = {
  "deceptive": "https://fatabyyano.net/newsface/deceptive/",
  "sarcastic": "https://fatabyyano.net/newsface/sarcastic/",
  "misleading-title":"https://fatabyyano.net/newsface/misleading-title/",
  "fake": "https://fatabyyano.net/newsface/fake/",
  "not-qualified": "https://fatabyyano.net/newsface/not-qualified/",
  "partially-fake":"https://fatabyyano.net/newsface/partially-fake/"
}
for x in pages:
  count=0
  csvfile = open('drive/My Drive/Arabic Fake news/Fatabayanoo/csvfiles/'+x+'.csv', 'w+', newline='')
  writer = csv.writer(csvfile)
  for i in range(15):
    URL = pages[x]+"page/"+str(i+1)+'/'
    page = requests.get(URL)
    soup = BeautifulSoup(page.content, 'html.parser')
    news_elems = soup.find_all('h2', class_='post_title')
    for elem in news_elems:
      href = elem.find('a')['href']
      if href:
        writer.writerow([href])
        count=count+1
  print(str(x)+".csv count = "+str(count))
  csvfile.close()

print("Finish")
```

```
from newspaper import Article
import csv
import json
import os

pages = {
  "deceptive": 'drive/My Drive/Arabic Fake news/Fatabayanoo/csvfiles/deceptive.csv',
  "sarcastic": 'drive/My Drive/Arabic Fake news/Fatabayanoo/csvfiles/sarcastic.csv',
  "misleading-title":'drive/My Drive/Arabic Fake news/Fatabayanoo/csvfiles/misleading-title.csv',
  "fake": 'drive/My Drive/Arabic Fake news/Fatabayanoo/csvfiles/fake.csv',
  "not-qualified": 'drive/My Drive/Arabic Fake news/Fatabayanoo/csvfiles/not-qualified.csv',
  "partially-fake":'drive/My Drive/Arabic Fake news/Fatabayanoo/csvfiles/partially-fake.csv'
}
for x in pages:
  basePath='drive/My Drive/Arabic Fake news/Fatabayanoo/RawData/'+str(x)+'/'
  if not os.path.exists(basePath):
    os.makedirs(basePath)
  i=0  
  file = open(pages[x])
  csv_file = csv.reader(file)
  for link in csv_file:
    article = Article(url=link[0], language='ar')
    article.download()
    if article.download_state != 1:
      i=i+1
      article.parse()
      news_dic = {
          'url': link[0],
          'title': article.title,
          'text': article.text,
          'img': article.top_image
          }
      with open(basePath+str(x)+str(i)+'.json','w',encoding='utf8') as outfile:
        json.dump(news_dic, outfile, indent=4,sort_keys=True, default=str,ensure_ascii=False)
  print("Finish Collecting " +str(x)+" "+str(i) +" news ")    
```

# Collect urls

```
#Get fake news url
import requests
from bs4 import BeautifulSoup
import codecs
from urllib.parse import unquote,quote
import csv

csvfile = open('drive/My Drive/Arabic Fake news/Akeed/notTrue.csv', 'w+', newline='')
writer = csv.writer(csvfile)
#writer.writerow(['Links'])
count=0
for i in range(100):
    URL ="https://akeed.jo/ar/mqal/4/%D8%A3%D8%AE%D8%A8%D8%A7%D8%B1_%D8%BA%D9%8A%D8%B1_%D8%B5%D8%AD%D9%8A%D8%AD%D8%A9/"+str(i+1)
    page = requests.get(URL)
    soup = BeautifulSoup(page.text, 'html.parser')
    news_elems = soup.find_all('div',class_="post-style4")
    for elem in news_elems:
      web_links = elem.find('a')['href']
      web_links="https://akeed.jo"+quote(web_links)
      if web_links:
        writer.writerow([web_links])
        count=count+1
print(count)   
csvfile.close()
print("Finish")
```

```
from newspaper import Article
import csv
import json
import os

akeedCSV='drive/My Drive/Arabic Fake news/Akeed/notTrue.csv'
basePath='drive/My Drive/Arabic Fake news/Akeed/RawData/'
if not os.path.exists(basePath):
  os.makedirs(basePath)

i=0
file = open(akeedCSV)
csv_file = csv.reader(file)

for link in csv_file:
  print(link[0])
  article = Article(url=link[0], language='ar')
  article.download()
  if article.download_state != 1:
    i=i+1
    article.parse()
    news_dic = {
        'url': link[0],
        'title': article.title,
        'text': article.text,
        'img': article.top_image
          }
    with open(basePath+"Akeed"+str(i)+'.json','w',encoding='utf8') as outfile:
      json.dump(news_dic, outfile, indent=4,sort_keys=True, default=str,ensure_ascii=False)
print("Finish")
print(i)
```

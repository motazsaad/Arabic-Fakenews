```
import requests
import csv
from bs4 import BeautifulSoup
from urllib.parse import unquote,quote

csvfile = open('drive/My Drive/Arabic Fake news/Verify/Verify.csv', 'w+', newline='')
writer = csv.writer(csvfile)
#file = open("drive/My Drive/Fake news Project/FakesURLs/VerifyUrls.txt", 'w',encoding='utf-8')
count=0
for i in range(100):
    URL ="https://www.verify-sy.com/list/16/%D9%85%D8%B9%D9%84%D9%88%D9%85%D8%A7%D8%AA-%D9%85%D8%B6%D9%84%D9%84%D8%A9?page="+str(i+1)
    page = requests.get(URL)
    soup = BeautifulSoup(page.content, 'html.parser')
    news_elems = soup.find_all('div',class_="blog_post_style2_content")
    #print(i,len(news_elems),'Collecting the links...')
    for elem in news_elems:
      href=elem.find('a')['href']
      if href:
        writer.writerow([href])
        count=count+1
print(count)   
csvfile.close()
print("Finish")
```

```
from newspaper import Article
import csv
import json
import os

verifyCSV='drive/My Drive/Arabic Fake news/Verify/Verify.csv'
basePath='drive/My Drive/Arabic Fake news/Verify/RawData/'
if not os.path.exists(basePath):
  os.makedirs(basePath)

i=0
file = open(verifyCSV)
csv_file = csv.reader(file)

for link in csv_file:
  article = Article(url=link[0], language='ar')
  article.download()
  if article.download_state != 1:
    i=i+1
    article.parse()
    news_dic = {
        'url': link[0],
        'title': article.title,
        'text': article.text,
        'img': article.top_image
        }
    with open(basePath+"verify"+str(i)+'.json','w',encoding='utf8') as outfile:
      json.dump(news_dic, outfile, indent=4,sort_keys=True, default=str,ensure_ascii=False)

print("Finish Collecting " +str(i) +" news ")    

  
```

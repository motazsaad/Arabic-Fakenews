```python
df_test = pd.read_csv('drive/My Drive/Arabic_FakeNews/Model/Balance/DataTitle/PPV2test_news.csv')
df_test.text=df_test.text.astype(str)
df_test.head()
```

```python
x_test = df_test.text
y_test = df_test.label
```

```python
from transformers import pipeline
Model='joeddav/xlm-roberta-large-xnli'
classifier = pipeline("zero-shot-classification", model=Model,tokenizer=Model)
```

```python
text_example = pd.read_csv("drive/My Drive/Arabic_FakeNews/Model/Balance/Data/text_example.csv")
title_example = pd.read_csv("drive/My Drive/Arabic_FakeNews/Model/Balance/Data/title_example.csv")
```

# First

## Sample Of Data

```python
subset_news=[]
for news in title_example.text:
  pred_zero=classifier(news, ["خبر حقيقي", "خبر كاذب"])
  SCORES = pred_zero["scores"][0]
  labels=pred_zero["labels"][0]
  print(SCORES)
  print(labels)
```

```python
subset_news=[]
for news in text_example.text:
  pred_zero=classifier(news, ["خبر حقيقي", "خبر كاذب"])
  SCORES = pred_zero["scores"][0]
  labels=pred_zero["labels"][0]
  print(SCORES)
  print(labels)
```

## All Data

```python
i=0
y_pred_new=[]
for news in df_test.text:
  i=i+1
  pred_zero=classifier(news, ["خبر حقيقي", "خبر كاذب"])
  if(pred_zero["scores"][0]> pred_zero["scores"][1]):
    y_pred_new.append(pred_zero["labels"][0])
  else:
    y_pred_new.append(pred_zero["labels"][1])
  
  print(str(i) +"-"+news)

y_pred_new[:5]
```

## Accuercy

```python
convert = lambda x : 0 if x=='خبر كاذب' else 1
```

```python
y_pred=[]
for news in y_pred_new:
  y_pred.append(convert(news))
```

```python
print("Accuracy", accuracy_score(y_test, y_pred))
print("f1_score", f1_score(y_test, y_pred))
print("\nConfusion matrix\n",confusion_matrix(y_test,y_pred))
print("\nClassification report\n",classification_report(y_test,y_pred))
```

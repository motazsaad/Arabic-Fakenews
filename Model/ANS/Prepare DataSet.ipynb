```python
import tensorflow as tf
import pandas as pd
from sklearn.model_selection import train_test_split
from arabert.preprocess import ArabertPreprocessor
```

```python
#Call Model
model_name="bert-base-arabertv02"
processor = ArabertPreprocessor(model_name=model_name)
```

```python
import pandas as pd
dev_df= pd.read_csv("/content/drive/MyDrive/Arabic_FakeNews/Model/AraNews/Data/wanlp2020_arabic_fake_news_detection-main/ArNews_dev_GW_order_0.tsv",index_col=0)
train_df= pd.read_csv("/content/drive/MyDrive/Arabic_FakeNews/Model/AraNews/Data/wanlp2020_arabic_fake_news_detection-main/ArNews_train_GW_order_0.tsv",index_col=0)
test_df= pd.read_csv("/content/drive/MyDrive/Arabic_FakeNews/Model/AraNews/Data/wanlp2020_arabic_fake_news_detection-main/ArNews_test_GW_order_0.tsv",index_col=0)
```

```python
train_df
```

```python
train_df.drop(columns=['SentID'],axis=1,inplace=True)
train_df.rename(columns = {'Sentence' : 'text'}, inplace = True)
train_df['label']=train_df['label'].apply(lambda x: 1 if x=='Orginal' else 0 )

train_df.head()
```

```python

dev_df.drop(columns=['SentID'],axis=1,inplace=True)
dev_df.rename(columns = {'Sentence' : 'text'}, inplace = True)
dev_df['label']=dev_df['label'].apply(lambda x: 1 if x=='Orginal' else 0 )

dev_df.head()
```

```python
test_df.drop(columns=['SentID'],axis=1,inplace=True)
test_df.rename(columns = {'Sentence' : 'text'}, inplace = True)
test_df['label']=test_df['label'].apply(lambda x: 1 if x=='Orginal' else 0 )

test_df.head()
```

```python
train_df.text = train_df.text.apply(lambda x:   processor.preprocess(x))
train_df.to_csv('/content/drive/MyDrive/Arabic_FakeNews/Model/AraNews/Data/PPV2train_news.csv', index=False)

dev_df.text = train_df.text.apply(lambda x:   processor.preprocess(x))
dev_df.to_csv('/content/drive/MyDrive/Arabic_FakeNews/Model/AraNews/Data/PPV2dev_news.csv', index=False)

test_df.text = test_df.text.apply(lambda x:   processor.preprocess(x))
test_df.to_csv('/content/drive/MyDrive/Arabic_FakeNews/Model/AraNews/Data/PPV2test_news.csv', index=False)
```
